```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
data_linear1 = {
    "temperature": [20, 22, 25, 28, 30],
    "sales": [200, 250, 300, 380, 420]
}
```

```python
df=pd.DataFrame(data_linear1)
x=df[["temperature"]]
y=df["sales"]
model=LinearRegression()
model.fit(x,y)
# temp=float(input("Enter temperature : "))
print(f"The sales when temperature is 12 :", model.predict([[12]]))
```
